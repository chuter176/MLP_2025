if_elif_else

```python
x = 1
y = 2
if x < y:
  print('x가 y보다 작습니다.')
else:
  print('x가 y보다 작지 않습니다.')
```

```python
x = 1
y = 2
if x < y:
  print('x가 y보다 작습니다.')
elif x > y:
  print('x가 y보다 큽니다.')
else:
  print('x와 y가 같습니다.')
```
